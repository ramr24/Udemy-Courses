```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.left = None #left child
        self.right = None #right child
    
class Tree:
    #1)Write a function that takes in a list of integers, creates a binary tree with those integers
    def __init__(self):
        self.root = None 

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert(data, self.root)
    
    #Helper method for insert     
    def _insert(self, data, cur_node):
        if data < cur_node.data:
            if cur_node.left is None:
                cur_node.left = Node(data)
            else: 
                self._insert(data, cur_node.left)
        elif data > cur_node.data:
            if cur_node.right is None:
                cur_node.right = Node(data)
            else:
                self._insert(data, cur_node.right)
        else:
            print("Value is already present in tree.")
    
    #2)Write a function that returns the in-order traversal of the tree as space-separated string.
    def in_order(self):
        print("In Order: Left -> Root -> Right")
        if self.root:
            self._in_order(self.root)
    
    #Helper method for in_order
    def _in_order(self, cur_node):
        if cur_node:
            self._in_order(cur_node.left)
            print(str(cur_node.data))
            self._in_order(cur_node.right)

    #3)Write a function that returns the pre-order traversal of the tree as space-separated string.
    def pre_order(self):
        print("Pre Order: Root -> Left -> Right")
        if self.root:
            self._pre_order(self.root)
    
    #Helper method for pre_order
    def _pre_order(self, cur_node):
        if cur_node:
            print(str(cur_node.data))
            self._pre_order(cur_node.left)
            self._pre_order(cur_node.right)
    
    #4)Write a function that returns the post-order traversal of the tree as space-separated string.
    def post_order(self):
        print("Post Order: Left -> Right -> Root")
        if self.root:
            self._post_order(self.root)
    
    #Helper method for post_order
    def _post_order(self, cur_node):
        if cur_node:
            self._post_order(cur_node.left)
            self._post_order(cur_node.right)
            print(str(cur_node.data))
            
    #5)Write a function that determines the height of a given tree.
    def height(self):
        if self.root is not None:
            return self._height(self.root, -1)
        else:
            return 0
        
    #Helper method for height
    def _height(self, cur_node, cur_height):
        if cur_node is None:
            return cur_height
        left_height = self._height(cur_node.left, cur_height + 1)
        right_height = self._height(cur_node.right, cur_height + 1)
        return max(left_height, right_height)            

    #7)Write a function that returns a bool indicating that a value exists (or not) in a given tree.
    def contains(self, data):
        if self.root:
            exist = self._contains(data, self.root)
            if exist:
                return True
            return False
        else:
            return None
        
    #Helper method for contains 
    def _contains(self, data, cur_node):
        if data > cur_node.data and cur_node.right:
            return self._contains(data, cur_node.right)
        elif data < cur_node.data and cur_node.left:
            return self._contains(data, cur_node.left)
        if data == cur_node.data:
            return True
```

```python
tree = Tree()
tree.insert(10)
tree.insert(15)
tree.insert(7)
tree.insert(9)
tree.insert(3)
tree.insert(24)

tree.in_order()
print()
tree.pre_order()
print()
tree.post_order()
print()
print(tree.contains(10))
print(tree.contains(15))
print(tree.contains(7))
print(tree.contains(9))
print(tree.contains(3))
print(tree.contains(24))
print(tree.contains(11))
tree.height()
```
