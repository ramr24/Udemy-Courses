## Drop  Rows with Null Values

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.tail(3)
```

```python
nba.dropna()
```

.dropna() drops rows that has the value NaN 

```python
nba.dropna(how = "all")
```

.dropna(how = "all") only drops rows that have NaN in all columns but keeps the data that have NaN in some columns

```python
nba.dropna(how = "all", inplace = True)
nba.tail(3)
```

```python
nba.dropna(axis = 1)
```

Remove any columns that have the value NaN

```python
nba.dropna(axis = "columns")
```

```python
nba.head(3)
```

```python
nba.dropna(subset = ["Salary"])
```

.dropna(subset = ["Column Name"]) drops the rows within the listed column that has NaN 

```python
nba.dropna(subset = ["Salary", "College"])
```

## Fill in Null Values with the .fillna() Method

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.fillna(0)
```

.fillna() fills the NaN value with the a value listed in the parentheses

You have to be careful and make sure which columns are numerical and categorical because .fillna() doesn't account for that on its own

```python
nba["Salary"].fillna(0, inplace = True)
nba.head()
```

```python
nba["College"].fillna("No College", inplace = True)
nba.head()
```

The inplace = True is used to modify the fill which will afffect the original data frame with the modified changes

## The .astype() Method 

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"].fillna(0, inplace = True)
nba["College"].fillna("None", inplace = True)
nba.head(6) 
```
