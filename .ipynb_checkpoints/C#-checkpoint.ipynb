```python
#C-sharp
# // = comments(same as #comment in python)
# /** = paragraph comments(same as """comments""" in python)
# 0x1 = hexadecimal (same as 1 in python)
# uint = like int but doesn't store negative numbers


#General
# num /= base; num = num / base
```

```python
#Assumes number >= 0 and base > 0; Does no validation of input
public static string Convert(number, baseValue)
        #Lets not worry about bases > 16, or any negative numbers.
    if ((number < 0) || (baseValue < 0) || (baseValue > 16))
    return "";

    numBits = sizeof(int) * 8;
    StringBuilder bitString = new StringBuilder(numBits); 
    #optional to pass in capacity.. just being efficient to prevent reallocations
    for (ii = 0; ii < numBits; ++ii)
        bitString.Append("0");
    index = 0;
    while (number > 0)
        lsb = number % baseValue;  
        #lsb will have values from 0 to (baseValue - 1)
        bitString[numBits - index - 1] = GetChar(lsb);

        number /= baseValue;   
        #number = number / baseValue;
        ++index;
    return bitString.ToString();

private static char GetChar(number)
    char ch = '0';
    if (number <= 9)
        ch = (char)(((int)'0') + number);
    else
        ch = (char)(((int)'A') + (number - 10));

    return ch;

#Assumes number >= 0. Does no validation of input
public static string ConvertToBinary(number)
    numBits = sizeof * 8;
    StringBuilder bitString = new StringBuilder(numBits); 
    #optional to pass in capacity.. just being efficient to prevent reallocations
    for (ii = 0; ii < numBits; ++ii)
        bitString.Append("0");
    for (ii = 0; ii < numBits; ++ ii)
        lsb = ((number >> ii) & 1);
        if (lsb == 0)
            bitString[numBits - ii - 1] = '0';
        else
            bitString[numBits - ii - 1] = '1';

        """
           Or could say:  bitString[numBits - ii - 1] = (lsb == 0) ? "0" : "1");
           or bitString[numBits - ii - 1] = (((number >> ii) & 0x1) == 0 ? "0" : "1");   
           Not preferable, as harder to understand
        """
    return bitString.ToString();
```

```python
var Base62 = (function(){                
  var table = '0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ';
  
  function _to10(num)
  {
    var limit = num.length;
    var res = 0;
    for (var i = 0; i < limit; i ++)
    {
      res = 62 * res + table.indexOf(num.charAt(i));
    }
    return res;  
  }
  
  function _toBase(num)
  {
    var r = num % 62;
    var res = table.charAt(r);
    var q = Math.floor(num / 62);
    while (q)
    {
      r = q % 62;
      q = Math.floor(q / 62);
      res = table.charAt(r) + res;
    }
    return res;
  }
  
  return {
    FromBase10: function()
    {
      var r = [];
      for (var i = 0; i < arguments.length; i ++)
      {
        var num = parseInt(arguments[i]);
        r.push(_toBase(num));
      }
      return r;
    },
    
    FromBase62: function()
    {
      var r = [];
      for (var i = 0; i < arguments.length; i ++)
      {
        var num = arguments[i].toString();
        if (num.length)
        {
          r.push(_to10(num));
        }
      }
      return r;    
    }
  } 
})();
```
